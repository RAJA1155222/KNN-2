Euclidean: sensitive to outliers and large-scale features; works better when features are continuous and distances in space matter.

Rule of thumb:

Use Manhattan for high-dimensional, sparse, or categorical features encoded as numbers.

Q2. Choosing the optimal value of k

Effects of k:

Small k → high variance, sensitive to noise (overfitting)

Large k → high bias, may underfit

Techniques to determine optimal k:

Cross-validation: Split data into training/validation folds; pick k with best validation accuracy/MSE.

Grid search: Test multiple k values systematically and compare performance metrics.

Q3. How the distance metric affects performance

Distance metric defines “neighborhood” in KNN.

Euclidean vs Manhattan: can lead to different nearest neighbors, especially in high-dimensional space.

When to choose which:

Euclidean → continuous, low-dimensional, dense features

Manhattan → high-dimensional, sparse, or when features represent independent counts

Other metrics (like Minkowski, Chebyshev) can be used for specialized cases.

Q5. Effect of training set size on KNN performance

Larger training set → better performance (more representative neighbors)

Smaller training set → higher variance, possible misclassification

Tradeoff: Larger datasets → slower prediction (KNN is lazy learning)

Techniques to optimize training size:

Use representative sampling for huge datasets

Use KD-Tree or Ball Tree to speed up neighbor search

Dimensionality reduction to reduce computational cost while keeping relevant info

Q6. Potential drawbacks of KNN and solutions

Drawbacks:

Computationally expensive at prediction (O(n))

Sensitive to irrelevant/noisy features

Curse of dimensionality → distances lose meaning

Sensitive to feature scale

Solutions:

Feature scaling (StandardScaler, MinMaxScaler)

Dimensionality reduction (PCA, LDA)

Distance weighting to reduce influence of distant neighbors

Efficient nearest-neighbor search structures (KD-Tree, Ball Tree)

Feature selection to remove irrelevant/noisy features
